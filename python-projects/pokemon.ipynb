**VISUALIZATION**

```python
pokemon_counts = df['Type 1'].value_counts()


colors = plt.cm.tab20(range(len(pokemon_counts)))


plt.figure(figsize=(8, 8))
plt.pie(pokemon_counts, labels=pokemon_counts.index, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Distribution of Pokemon by type')
plt.axis('equal')


plt.show()



```

** Our graph represents a pie chart illustrating the distribution of Pokmon types. It shows the percentage of each Pokmon type in the dataset. Various types are represented with different colors. Among the most common types are 'Grass', 'Water', and 'Normal', while 'Flying' and 'Fairy' types are less frequent. The chart provides a quick glance at the relative distribution of Pokmon types within the Pokmon universe.**
