# Dependencies

```python
import pandas as pd
import numpy as np
```

```python
path = '/kaggle/input/airline-customer-satisfaction/Airline_customer_satisfaction.csv'
df = pd.read_csv(path)
df.head(5).transpose()
```

```python
# checking for missing values
df.isnull().sum()
```

```python
missing_values = df.loc[pd.isnull(df['Arrival Delay in Minutes'])]
```

```python
missing_values
```

```python
#deleting missin values
df.dropna(inplace=True)
```

```python
len(df)
```

```python
df.head(3)
```

```python
df['satisfaction'].value_counts()
```

```python
df = pd.get_dummies(df, columns=['Customer Type','Type of Travel','Class'], drop_first=True)
df['satisfaction'].value_counts()
```

```python
satisfaction_map = {'satisfied': 1, 'dissatisfied': 0}

df['satisfaction'] = df['satisfaction'].map(satisfaction_map)
```

```python
df['satisfaction']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
import xgboost as xgb
```

```python
X = df.drop('satisfaction',axis=1)
y = df['satisfaction']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_scaled, y_train)
```

```python
accuracy = rf_model.score(X_test_scaled, y_test)
print("Accuracy:", accuracy)
```

```python
gb_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
gb_model.fit(X_train_scaled, y_train)
```

```python
xgb_model = xgb.XGBClassifier(objective='multi:softmax', num_class=2, random_state=42)
xgb_model.fit(X_train_scaled, y_train)
```

```python
rf_pred = rf_model.predict(X_test_scaled)
gb_pred = gb_model.predict(X_test_scaled)
xgb_pred = xgb_model.predict(X_test_scaled)


rf_accuracy = accuracy_score(y_test, rf_pred)
gb_accuracy = accuracy_score(y_test, gb_pred)
xgb_accuracy = accuracy_score(y_test, xgb_pred)

print("Random Forest Accuracy Score:", rf_accuracy)
print("Gradient Boosting Accuracy Score:", gb_accuracy)
print("XGB Accuracy Score:", xgb_accuracy)
```
